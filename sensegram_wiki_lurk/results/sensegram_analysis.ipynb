## Анализ окрестностей смыслов слов (Sensegram) в Википедии и Луркморе

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

В этом ноутбуке мы рассмотрим, как распределены различные близости различных смыслов слов, а также пронаблюдаем, какие смыслы оказались наиболее близки и наиболее далеки в каждом случае

```python
results = pd.read_csv('results.csv', index_col=0)
```

Посмотрим, как соотносятся первые смыслы Википедии и Луркмора

```python
cosine_11 = results['Neighbours cosine dist Wiki1 -> Lurk1'].values
cosine_11 = cosine_11[~np.isnan(cosine_11)]
print('Sense mean: %1.3f' % np.mean(cosine_11))
print('Sense median: %1.3f' % np.median(cosine_11))
print('Sense variance: %1.3f' % np.var(cosine_11))
plt.hist(cosine_11, bins=100)
plt.show()
```

Видим, что распределение близостей имеет похожую на экспоненциальной природу. Значительное число слов имеют близость 0.

```python
%pylab inline

cols = ['Lurkmore first sense neighbours', 'Lurkmore second sense neighbours']
df = results.loc[results['Neighbours cosine dist Wiki1 -> Lurk1'] == 0][cols]
display(df)
```

Некоторые интересные кейсы: <b>"авангард"</b> (в википедии как хоккейный клуб, в луркморе как что-то передовое); <b>"японка"</b> (японская женщина и японская машина); <b>"адрес"</b> (почтовый и электронный); <b>"аквариум"</b> (как собственно аквариум и рок-группа); <b>"членистоногое"</b> (как тип первичноротых в википедии и понятно что в луркморе) и.т.д.

Теперь для уверенности посмотрим, какие слова наиболее близки

```python
df = results.loc[results['Neighbours cosine dist Wiki1 -> Lurk1'] > results['Neighbours cosine dist Wiki1 -> Lurk1'].quantile(.99)][cols]
display(df)
```

Видим, что наиболее похожи базовые однозначные термины вроде месяцев, что логично.
